## Data
- This mdata mined from google trends for each country with that countries language

```python
df = pd.read_csv("/Users/umar/Desktop/makale/cancer_goog_trends.csv")
df['Week'] = pd.to_datetime(df['Week'])
df['year_month'] = df['Week'].dt.to_period('M')
df['month'] = df['Week'].dt.month
df['year'] = df['Week'].dt.year
df.columns = ['week', 'breast_cancer', 'prostate_cancer', 'lung_cancer', 'country','year_month', 'month', 'year']
df.sample()
```

### monthly trends for each cancer type for each country

```python
df_breast_m = df.sort_values(by = ['week', 'country']).groupby(['country', 'month'])['breast_cancer'].mean().reset_index().pivot_table(index = 'month', columns = 'country', values = 'breast_cancer')
df_prostate_ = df.sort_values(by = ['week', 'country']).groupby(['country', 'month'])['lung_cancer'].mean().reset_index().pivot_table(index = 'month', columns = 'country', values = 'lung_cancer')
df_lung_m = df.sort_values(by = ['week', 'country']).groupby(['country', 'month'])['prostate_cancer'].mean().reset_index().pivot_table(index = 'month', columns = 'country', values = 'prostate_cancer')
```

### year monthly trends for each cancer type for each country

```python
df_breast_ym = df.groupby(['year_month', 'country'])['breast_cancer'].mean().reset_index().pivot_table(index = 'year_month', columns = 'country', values = 'breast_cancer')
df_prostate_ym = df.groupby(['year_month', 'country'])['lung_cancer'].mean().reset_index().pivot_table(index = 'year_month', columns = 'country', values = 'lung_cancer')
df_lung_ym = df.groupby(['year_month', 'country'])['prostate_cancer'].mean().reset_index().pivot_table(index = 'year_month', columns = 'country', values = 'prostate_cancer')
```
